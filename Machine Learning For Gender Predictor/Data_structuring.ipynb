```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image
%matplotlib inline
```

```python
###Image cropping
from glob import glob
female_path = glob('./data/female/*.jpg')
```

```python
male_path = glob('./data/male/*.jpg')
```

```python
len(female_path)
```

```python
len(male_path)
```

```python
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
```

```python
def extract_face(path, gender, i):
    img = cv2.imread(path)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = haar.detectMultiScale(gray,1.5,5)
    for x,y,w,h in faces:
        roi = img[y:y+h,x:x+w]
        if gender == 'male':
            cv2.imwrite('./data/crop/male_crop/{}_{}.png'.format(gender,i),roi)
        else:
            cv2.imwrite('./data/crop/female_crop/{}_{}.png'.format(gender,i),roi)
```

```python
for i,path in enumerate(female_path):
    try:
        extract_face(path,'female',i)
        print('INFO: {}/{} processed sucessfully'.format(i,len(female_path)))
    except:
        print('INFO: {}/{} cannot be processed'.format(i,len(female_path)))
```

```python
for i,path in enumerate(male_path):
    try:
        extract_face(path,'male',i)
        print('INFO: {}/{} processed sucessfully'.format(i,len(male_path)))
    except:
        print('INFO: {}/{} cannot be processed'.format(i,len(male_path)))
```

```python
def resize_img(path):
    img = cv2.imread(path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    if gray.shape[0] < 100:
        img = cv2.resize(gray,(100,100), cv2.INTER_AREA)##INTER_CUBIC for enlarge
    else:
        img = cv2.resize(gray,(100,100), cv2.INTER_CUBIC)
    flat_img = img.flatten()
        #     plt.imshow(img,cmap ='gray')
    return flat_img
```
